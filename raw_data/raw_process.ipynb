```python
"""
Pickle2JSON is a simple Python Command Line program for converting Pickle file to JSON file.
Arguments: Only one (1) argument is expected which is the pickle file.
Usage: python pickle2json.py myfile.pkl
Output: The output is a JSON file bearing the same filename containing the JSON document of the converted Pickle file.
"""

# import libraries
import pickle
import json
import sys
import os

# open pickle file
with open('datasets/IEMOCAP/new_dev_por_v4_0610_6c_sep_contexts.pkl', 'rb') as infile:
    obj = pickle.load(infile)

# convert pickle object to json object
json_obj = json.loads(json.dumps(obj, default=str))

# write the json file
with open(
        'iemocap_0610_sep_context.json',
        'w',
        encoding='utf-8'
    ) as outfile:
    json.dump(json_obj, outfile, ensure_ascii=False, indent=4)
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass

    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass

    return False

# def pre_gen(results):
#             #保障生成格式
#             new_results = []
#             for ele in results:
#                 if len(str(ele).split(','))==1:
#                     if is_number(str(ele)):
#                         new_results.append(str(ele)+','+'neutral')
#                     else:
#                         new_results.append('0.0'+','+str(ele))
#                 else:
#                     new_results.append(ele)
                    
#             return new_results

def pre_gen(results):
            #保障生成格式
            new_results = []
            for ele in results:
                if len(str(ele).split(','))==1:
                    if is_number(str(ele)):
                        new_results.append(str(ele)+','+'neutral,neutral,neutral')
                    else:
                        new_results.append('0.0'+','+str(ele) +',neutral,neutral')
                elif len(str(ele).split(','))==2:
                    new_results.append(str(ele)+','+'neutral' +',' + 'neutral')
                elif len(str(ele).split(','))==3:
                    new_results.append(str(ele)+','+'neutral')
                else:
                    new_results.append(ele)
                    
            return new_results

results = ['0.0', 'positive,joy', 'positive,0.0,joy', 'positive,0.0,joy,neutral']
new_res = pre_gen(results)
print(new_res)
```

```python
import pandas as pd 
data = pd.read_csv('datasets/IEMOCAP/IEMOCAP/new-iemocap-label-v4-contexts.csv')
for i in range(data.shape[0]):
    line = data.iloc[i]
    label = line['label']
    if label == 'neu':
        data.at[i, 'por'] = 'neutral'
    elif label in ['hap', 'exc']:
        data.at[i, 'por']= 'positive'
    else:
        data.at[i, 'por'] = 'negative'
data.to_csv('datasets/IEMOCAP/IEMOCAP/new-iemocap-label-v5-contexts.csv')
```
